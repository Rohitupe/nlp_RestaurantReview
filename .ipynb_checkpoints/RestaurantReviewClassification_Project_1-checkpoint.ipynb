```python
# Group above code in to one
import re
# Stop words not to consider
custom_stopwords = {"don","dont't","ain","aren","aren't","couldn","couldn't",
                   "didn","didn't","doesn","doesn't","hadn","hadn't","hasn","hasn't",
                   "haven","haven't","isn","isn't","ma","mightn","mightn't","mustn","mustn't",
                   "needn","needn't","shan","shan't","no","nor","not","shouldn","shouldn't",
                   "wasn","wasn't","weren","weren't","won","won't","wouldn","wouldn't"}
corpus = []

ps = PorterStemmer()
stop_words = set(stopwords.words("english")) - custom_stopwords

for i in range(len(data)):
    review = re.sub(r"[^a-zA-Z]", " ", data["Review"][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stop_words]
    review = " ".join(review)
    corpus.append(review)
```

```python
data["processed_text"] = corpus
```

```python
# Target variable here to analyze the review is "Liked" which makes it a binary classification problem as the values we have in the Liked are 0 & 1,  
# so here we will use difference ML classification algorithm, to classify different reviews,

# In order to do so we need to convert text data into numbers, and for that we will use bagOfWords technique(count vectorizer), as count vectorizer 
# converts text into numerical format
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2500) # so it include 1500 most recent words in our corpur
```

```python
X = cv.fit_transform(corpus).toarray()  # features --- independant varaible
```

```python
X
```

```python
X.shape
import joblib
joblib.dump(cv, "count_v_res")
```

```python
y = data["Liked"]  # dependant variable
```

```python
y
```

```python
# Apply Classification Algo, for train and test
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB  # First Model
```

```python
nb = GaussianNB()
nb.fit(X_train,y_train)
y_pred = nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)  # First model Accuracy Score
```

```python
from sklearn.linear_model import LogisticRegression # Second Model
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
accuracy_score(y_test, y_pred)  # Second Model accuracy score
```

```python
from sklearn.ensemble import RandomForestClassifier  # third model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test, y_pred)  # third model accuracy score, using this model for the model prediction lets save this model here
```

```python
# saving model
import joblib
joblib.dump(rf,'Restaurant_review_model')
```

```python
# Using the Ml Model with processed text as input
import joblib
import numpy as np
input_data = "not tasti textur nasti"
model = joblib.load("Restaurant_review_model")
vec = joblib.load("count_v_res")
processed_input_vec = vec.transform([input_data])
prediction = model.predict(processed_input_vec)[0]
print(prediction)
# if 1 -- positive review, if 0 -- negative review
```
